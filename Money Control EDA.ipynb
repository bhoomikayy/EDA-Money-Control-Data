### Trends Followed by the Dataset

```python
#The following function defines a method to study the trend of each stock in terms of %Change
#The data has been categorised into 9 categories and a graph has been plotted for the same.

def trend(x):
  if x > -0.5 and x <= 0.5:
    return 'Slight or No change'
  elif x > 0.5 and x <= 1:
    return 'Slight Positive'
  elif x > -1 and x <= -0.5:
    return 'Slight Negative'
  elif x > 1 and x <= 3:
    return 'Positive'
  elif x > -3 and x <= -1:
    return 'Negative'
  elif x > 3 and x <= 7:
    return 'Among top gainers'
  elif x > -7 and x <= -3:
    return 'Among top losers'
  elif x > 7:
    return 'Bull run'
  elif x <= -7:
    return 'Bear drop'

df['Trend']= np.zeros(df['%Change'].count())

df['Trend']= df['%Change'].apply(lambda x:trend(x))
df
```

```python
#Count for each category of trend a particular company stock is following

group_names=['Slight or No change','Slight Positive','Slight Negative','Positive','Negative','Among top gainers','Among top losers','Bull run','Bear drop'  ]
df[['%Change','Trend']].head(20)
df["Trend"].value_counts()
```

### Univariate Categorical Data Analysis

```python
#Univariate Categorical Analysis: Here we study individual characteristics of every feature/variable available in the dataset.

def UVA_Category(data, var):
    size=len(var)
    plt.figure(figsize=(8*size,7), dpi=100)
    
    for j,i in enumerate(var):
        norm_count=data[i].value_counts(normalize=True)
        n_uni=data[i].nunique()
        
        plt.subplot(1,size,j+1)
        graph2=sns.countplot(y=i,data=data,order=data[i].value_counts().index)
        for p in graph2.patches:
            graph2.annotate(s='{:.0f}'.format(p.get_width()),xy=(p.get_width()+0.1,p.get_y()+0.7))     
```
